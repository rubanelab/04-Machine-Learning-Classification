### LogisticRegression (With GV)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# getting ERROR when passing the 'penalties': ['12'], so I removed
params = {
            'solver': ['newton-cg', 'lbfgs', 'saga']
         }

grid = GridSearchCV(LogisticRegression(), params, refit=True, verbose=3, n_jobs=-1, scoring="f1_weighted")
grid.fit(X_train, y_train)
```

### SVM Classifer (With CV)

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# getting ERROR when passing the 'penalties': ['12'], so I removed
params = {
                "kernel": ["linear", "rbf", "sigmoid"],
                "C": [10, 100, 1000],
                "gamma": ["auto", "scale"]
         }


grid = GridSearchCV(SVC(probability=True), params, refit=True, verbose=3, n_jobs=-1, scoring="f1_weighted")
grid.fit(X_train, y_train)
```

```python
results = grid.cv_results_
print(results)
```

```python
grid_pred = grid.predict(X_test)
grid_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

matrix = confusion_matrix(y_test, grid_pred)
matrix
```

```python
report = classification_report(y_test, grid_pred)
print(report)
```

### Decision Tree Classifier (CV)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

params = {
        "criterion": ['gini', 'entropy', 'log_loss'],
        "splitter": ["best", "random"]
       }

grid = GridSearchCV(DecisionTreeClassifier(), params, refit=True, verbose=3, n_jobs=-1, scoring="f1_weighted")
grid.fit(X_train, y_train)
```

```python
results = grid.cv_results_
print(results)
```

```python
grid_pred = grid.predict(X_test)
grid_pred
```

```python
matrix = confusion_matrix(y_test, grid_pred)
matrix
```

```python
report = classification_report(y_test, grid_pred)
print(report)
```
