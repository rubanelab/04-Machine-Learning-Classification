```python
# Import required libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype=int, drop_first = True)
```

```python
dataset
```

```python
dataset.columns
```

```python
# Since the User_ID column doesn't give meaningful info, we drop the data in rows (axis = 1 means row)
dataset = dataset.drop("User ID", axis=1)
dataset
```

```python
# Check the values counts. It will help to check the dataset balanced or not
dataset["Purchased"].value_counts()
```

```python
X = dataset[["Age", "EstimatedSalary", "Gender_Male"]]
y = dataset["Purchased"]
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# getting ERROR when passing the 'penalties': ['12'], so I removed
params = {
                "kernel": ["linear", "rbf", "sigmoid"],
                "C": [10, 100, 1000, 2000],
                "gamma": ["auto", "scale"]
         }


grid = GridSearchCV(SVC(probability=True), params, refit=True, verbose=3, n_jobs=-1, scoring="f1_weighted")
grid.fit(X_train, y_train)
```

```python
results = grid.cv_results_
print(results)
```

```python
grid_pred = grid.predict(X_test)
grid_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
matrix = confusion_matrix(y_test, grid_pred)
```

```python
print(matrix)
```

```python
report = classification_report(y_test, grid_pred)
print(report)
```

```python
from sklearn.metrics import f1_score
f1_macro = f1_score(y_test, grid_pred, average="weighted")
print("The f1_macro value for best parameter {}:".format(grid.best_params_), f1_macro)
```

```python
from sklearn.metrics import roc_auc_score

# It will not work, if not provide the probability data (predict_proba)
roc_auc_score(y_test, grid.predict_proba(X_test)[:,1])
```
